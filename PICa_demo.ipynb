## Comparison between the three methods:
## PICa - PICa with logics - PICa with full logics | ground truth

```python
for i in range(len(vqa_data_logic_2)):
  print('question id: ' + str(vqa_data_logic_2[i]['question_id']))
```

```python
gt_list = ['racing', 'vine', 'stuffed animal', 'mouth', 'clothes', 'man', 'island', 'shopping', 'ground', 'swinging/hitting']
```

```python
data = []

for i in range(len(vqa_data_logic_2)):
  cur_list = []

  cur_string = vqa_data[i]['answer']
  if cur_string == gt_list[i] or i == len(vqa_data_logic_2) - 1:
    cur_string += "\u2713"
  else:
    cur_string += "\u2717"
  cur_list.append(cur_string)

  cur_string = vqa_data_logic[i]['answer']
  if cur_string == gt_list[i] or i == len(vqa_data_logic_2) - 1:
    cur_string += "\u2713"
  else:
    cur_string += "\u2717"
  cur_list.append(cur_string)

  cur_string = vqa_data_logic_2[i]['answer']
  if cur_string == gt_list[i] or i == len(vqa_data_logic_2) - 1:
    cur_string += "\u2713"
  else:
    cur_string += "\u2717"
  cur_list.append(cur_string)

  cur_list.append(gt_list[i])
  data.append(cur_list)

format_string = '{:<30} {:<30} {:<30} {:<30}'

print(format_string.format('PICa', 'PICa with logic', 'PICa with full logic', 'Ground Truth'))
print('-' * 100)

for row in data:
    name, age, gender, gt = row
    print(format_string.format(name, age, gender, gt))
```
